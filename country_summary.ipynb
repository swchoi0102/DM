```python
data = pd.read_csv("sales9.csv")
```

```python
countries = data["COUNTRY"].unique()
```

```python
with open("country_summary.csv", "w", encoding = "utf-8") as f:
    
    f.write("country, 총매출액, 건당 평균 구매액, 구매횟수\n")
    
    for i in countries:
        print (i + " start!\n")
        sum_ = data.ix[data['COUNTRY'] == i].sum()[4]
        print (i + " sum end!\n")
        count_ = data.ix[data['COUNTRY'] == i].count()[4]
        print (i + " count end!\n")
        mean_ = sum_/count_
        print (i + " mean end!\n")
        
        line = i + ", " + str(sum_) + ", " + str(mean_) + ", " + str(count_) + "\n"
        
        f.write(line)
```
