```python
import pandas as pd
```

```python
data = pd.read_csv("sales8.csv")
data.head()
```

```python
del data["Unnamed: 0"] # 필요없는 열 삭제
```

```python
data.head()
```

```python
countries = data['COUNTRY'].unique()
countries
```

```python
# 3-Letter 나라 코드들로 바꿀 2-Letter dictionary
convert_country = {
    "CN ":"CHN", "KR ":"KOR", "JP ":"JPN", "VIE":"VNM", "SIN":"SGP",
    "THI":"THA", "US ":"USA", "HKO":"HKG", "TW ":"TWN", "HK ":"HKG",
    "TH ":"THA", "PH ":"PHL",
}

# replace country names 2-Letter -> 3-Letter
for old_country_name, new_country_name in convert_country.items():
    data["COUNTRY"].replace([old_country_name],[new_country_name], inplace=True)

countries = data['COUNTRY'].unique()
countries
```

```python
delete_country = {
    "ZZ ":0
}

for c in countries:
    count = data.loc[data['COUNTRY'] == c].count()[0] # count 
    if count < 120: delete_country[c] = count         # 거래건수가 120 이하인 나라들은 삭제

delete_country # 삭제할 나라 목록과 거래건수
```

```python
delete_country.keys() # 삭제할 나라 목록
```

```python
data = data[data["COUNTRY"].isin(delete_country.keys()) == False]  # 삭제
data['COUNTRY'].unique()  # 삭제한 뒤에 남은 나라들
```
